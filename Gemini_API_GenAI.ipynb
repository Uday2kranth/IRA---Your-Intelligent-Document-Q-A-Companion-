### Text Bot

```python
txt_history = [
    {'role' : 'user', 'parts' : 'hello'},
    {'role' : 'model', 'parts' : 'Great to meet you, what do you want me to do?'}
]

txt_chat_model = text_model.start_chat(
    history = txt_history
)

user_input = input("Enter the prompt: ")

while user_input!='close':
  response_txt_chat = txt_chat_model.send_message(user_input)
  display(Markdown(response_txt_chat.text))

  user_append = {
      'role' : 'user',
      'parts' : user_input
  }

  bot_append = {
      'role' : 'model',
      'parts' : response_txt_chat.text
  }

  txt_history.append(user_append)
  txt_history.append(bot_append)

  user_input= input("Enter the prompt: ")
```

### Generative Configuration and Streaming

```python
model_gen_config = genai.GenerativeModel(
    model_name = "gemini-1.5-flash",
    system_instruction = "You are a very talented doctor, who has an experience of over 27 years, in the domain of neurology",
    generation_config = genai.GenerationConfig(
        temperature = 0.7,
        top_p = 0.96,
        top_k = 2,
        candidate_count = 1,
        max_output_tokens = 1024
    )
)
```

```python
response_stream = model_gen_config.generate_content("Explain about Quantum machine Learning", stream = True)
```

```python
for chunk in response_stream:
  display(Markdown(chunk.text))
  print("-----------------------")
```

### Document Chat

```python
doc_path = "/content/Hybrid Classical-Quantum Steiner Tree Optimization.pdf"
```

```python
doc_model = genai.GenerativeModel(
    model_name = 'gemini-2.0-flash',
    system_instruction = "you are a proficient text analyzer, who can go through the text of the provided file and answer the queries"
)
```

```python
pdf_file = genai.upload_file(
    doc_path,
    mime_type = 'application/pdf'
)
```

```python
doc_history = [
    {'role' : 'user', 'parts' : 'hello'},
    {'role' : 'model', 'parts' : 'Great to meet you, what do you want me to do?'}
]

doc_chat_model = doc_model.start_chat(
    history = doc_history
)

user_input = input("Enter the prompt: ")

while user_input!='close':
  response_doc_chat = doc_chat_model.send_message([user_input, pdf_file])
  display(Markdown(response_doc_chat.text))

  user_append = {
      'role' : 'user',
      'parts' : user_input
  }

  bot_append = {
      'role' : 'model',
      'parts' : response_doc_chat.text
  }

  doc_history.append(user_append)
  doc_history.append(bot_append)

  user_input= input("Enter the prompt: ")
```
